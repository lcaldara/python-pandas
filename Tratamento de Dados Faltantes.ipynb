## Tratamento de Dados Faltantes (Condomínio)

```python
dados[dados['Condominio'].isnull()]
```

```python
selecao = (dados['Condominio'].isnull()) & (dados['Tipo'] == 'Apartamento')
selecao.shape[0]
```

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape [0]
A - B
```

```python
dados[dados['Condominio'].isnull()]
```

```python
dados.info()
```

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```
